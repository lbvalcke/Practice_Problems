```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the countTriplets function below.
def countTriplets(arr, r):
    trip_dict = {}
    triplets = 0
    for index, value in enumerate(arr):
        if value in trip_dict:
            trip_dict[value].append(index)
        else:
            trip_dict[value] = [index]
    for key,value in trip_dict.items():
        if (key * r) in trip_dict and (key * r * r) in trip_dict:
            for i in trip_dict[key]:
                for j in trip_dict[key * r]:
                    if i < j:
                        for k in trip_dict[key * r * r]:
                            if j < k:
                                triplets += 1
                            else:
                                pass
                    else:
                        pass
        else:
            pass
    
    return triplets

    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    nr = input().rstrip().split()

    n = int(nr[0])

    r = int(nr[1])

    arr = list(map(int, input().rstrip().split()))

    ans = countTriplets(arr, r)

    fptr.write(str(ans) + '\n')

    fptr.close()
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the freqQuery function below.
def freqQuery(queries):
    num_dict={}
    freq_dict={}
    results_array = []
    for query in queries:
        if query[0] == 1:
            num_dict.setdefault(query[1],0)
            freq_dict.setdefault(num_dict[query[1]],0)

            freq_dict[num_dict[query[1]]] -= 1
            num_dict[query[1]] += 1            
            
            freq_dict.setdefault(num_dict[query[1]],0)
            freq_dict[num_dict[query[1]]] += 1 
            
        elif query[0] == 2:
            num_dict.setdefault(query[1],0)
            if num_dict[query[1]] >= 1:
                freq_dict[num_dict[query[1]]] -= 1
                num_dict[query[1]] -= 1
                freq_dict[num_dict[query[1]]] += 1 
            else:
                pass
        elif query[0] == 3:
            freq_dict.setdefault(query[1],0)
            if freq_dict[query[1]] >= 1:
                results_array.append(1)
            else:
                results_array.append(0)
    return results_array

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    q = int(input().strip())

    queries = []

    for _ in range(q):
        queries.append(list(map(int, input().rstrip().split())))

    ans = freqQuery(queries)

    fptr.write('\n'.join(map(str, ans)))
    fptr.write('\n')

    fptr.close()
```
